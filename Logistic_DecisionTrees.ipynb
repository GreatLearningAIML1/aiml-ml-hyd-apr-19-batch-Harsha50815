### Divide the dataset into feature set and target set

```python
y=bc2['Class']
X=bc2.drop(columns='Class')
```

### Divide the Training and Test sets in 70:30 

```python
X_train, X_test, train_labels, test_labels = train_test_split(X, y, test_size=.30, random_state=1)
```

### Model 1 : Implement a default Decision Tree model with no attributes

```python
from sklearn.tree import DecisionTreeClassifier
dtreg = DecisionTreeClassifier()
dtreg.fit(X_train,train_labels)
```

### Check the feature importances of the Decision Tree model

```python
feature_importances = dtreg.feature_importances_
feature_names = bc2.columns
print(feature_names,feature_importances)
```

### Check the training & test accuracy of model 1 

```python

print(dtreg.score(X_train , train_labels))
print(dtreg.score(X_test , test_labels))
```

### Model 2: Implement a decision tree with depth of the tree = 10 and leaf nodes = 5

```python
dtreg2 = DecisionTreeClassifier(max_depth=10, max_leaf_nodes=5)
dtreg2.fit(X_train,train_labels)
```

### Check the training & test accuracy of model 2

```python

print(dtreg2.score(X_train , train_labels))
print(dtreg2.score(X_test , test_labels))
```

### Print the confusion matrix of model 1 and model 2

```python
from sklearn.metrics import confusion_matrix
y1_pred=dtreg.predict(X_test)
confusion_matrix(test_labels,y1_pred)
```

```python
y2_pred=dtreg2.predict(X_test)
confusion_matrix(test_labels,y2_pred)
```
