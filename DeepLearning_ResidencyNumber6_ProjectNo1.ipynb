1.Read the dataset

```
import numpy as np
import pandas as pd
bankdata = pd.read_csv('/content/bank.csv')
bankdata
```

```
bankdata.info()
```

```
bankdata.isna()
```

2. Drop the columns which are unique for all users like IDs (2.5 points)

```
bankdata = bankdata.drop(['Surname','CustomerId','RowNumber'], axis=1)
bankdata
```

Distinguish the feature and target set (2.5 points)


```
features = bankdata.drop(['Exited'], axis =1)
target = bankdata['Exited']
features.info()
```

```
bankdata.Exited.value_counts()
```

Label encoding for object data

```
from sklearn import preprocessing

labelencode1 = preprocessing.LabelEncoder()
labelencode2 = preprocessing.LabelEncoder()

features["Geography"] = labelencode1.fit_transform(features["Geography"])
features["Gender"] = labelencode2.fit_transform(features["Gender"])
```

```
features.info()
```

4.Divide the data set into Train and test sets

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3,random_state = 123)
```

Normalize the train and test data (2.5 points)


```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train.shape
```

Using Tenserflow

```
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
#Initialize Sequential Graph (model)
model = tf.keras.Sequential()

#Add Dense layer for prediction - Keras declares weights and bias automatically
model.add(tf.keras.layers.Dense(1, input_shape=(10,)))

#Compile the model - add Loss and Gradient Descent optimizer
model.compile(optimizer='sgd', loss='mse')
```

```
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Fitting the model

```
model.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

Predict the results using 0.5 


```
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

Accuracy score and confusion matrix

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```
cmatrix= confusion_matrix(y_pred, y_test)
cmatrix
```

```
classification_report(y_pred, y_test)
```
