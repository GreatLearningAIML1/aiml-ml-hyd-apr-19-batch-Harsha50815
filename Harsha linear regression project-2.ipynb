```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_table("http://www.stat.ufl.edu/~winner/data/airq402.dat", sep="\s+", header = None)
```

```python
dataset
```

```python
dataset.columns = ['city1','city2','Average_Fare',
                     'Distance','Average_weekly_passengers','market_leading_airline','market_share','Average_fare','Low_price_airline','market_share','price']
```

```python
dataset.describe()
```

```python
# 2.Remove the outliers from the data (5 points)
```

```python
dataset.plot.box(grid='True')
```

```python
dataset.head()
print('Shape of Dataset Before removing the outliers:',dataset.shape)
n_columns = data.select_dtypes(include=[np.number]).columns.tolist()

n_columns

for cl in n_columns:
    sorted_col = data.sort_values([cl])
    q1, q3= np.percentile(sorted_col[cl],[25,75])
    print(q1,q3)
    iqr = q3 - q1
    lower = q1 -(1.5 * iqr) 
    upper = q3 +(1.5 * iqr) 
    data = data.loc[(data[cl] > lower) & (data[cl] < upper)]
```

```python
#6. Divide the data set into training and test data set and build a Multiple Linear Regression model
```

```python
dataset2=dataset1.drop(['city1','city2','market_leading_airline','Low_price_airline'], axis=1)
dataset2.head()
dataset2.describe()
```

```python
dataset3=dataset2.values
dataset3=dataset3[:,1:]
dataset3.shape
```

```python
X_dataset=dataset3[:,1:]# from first column to all , with all rows 
Y_dataset=dataset3[:,0:1]#
print(X_dataset,Y_dataset)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dataset, Y_dataset, test_size=0.20)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
#7.Print the coefficients & intercepts of the linear regression model
```

```python
from sklearn.linear_model import LinearRegression  
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
a=lr.intercept_
b=lr.coef_
```

```python
y=a+X_train+b
y
```

```python
y_pred = lr.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

mse = mean_squared_error(y_test, y_pred)

rmse = sqrt(mse)
rmse
```

```python
# 8. Print the accuracy of the overall model
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
