```python
os=np.array(ticData1['Old Scheme (in thousands)'])
ns=np.array(ticData1['New Scheme (in thousands)'])
```

```python
from scipy.stats import shapiro,levene,ttest_1samp
shapiro(os)
shapiro(ns)
levene(os,ns)
```

```python
tstat,pval=ttest_1samp(ns-os,0)
pValue = round(pval/2,3)
pValue
```

```python
#p value = 0.065 > 0.05 
#H0 should not be rejected.

for i in range(11):
    ttes,pval=ttest_1samp(ns-os,i)
    print("diffrence of {} is {} : Confidence {}% and reject H0 but not 95% ".format(i,round(pval,2),round(1-pval,2)))
```
